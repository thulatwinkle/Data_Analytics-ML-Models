```python
#Load libraries

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import input file
# Reshape numpy objects to get 2D array so that it can fit to model
data = pd.read_csv('../input/canada_per_capita_income.csv')
x = data.iloc[:,0].values.reshape(-1,1)
y = data.iloc[:,1].values.reshape(-1,1)

```

```python
#Visualize data
data.head()
sns.heatmap(data.corr())
```

```python
#Train and Test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
#Train model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Test model
y_pred = lr.predict(x_test)
y_pred
```

```python
#Visualize training dataset result
plt.scatter(x_train,y_train,color='blue')  #Actual values
plt.plot(x_train,lr.predict(x_train),color='red') #Predict value using linear regression
plt.title('To find Per Capita Income')
plt.xlabel('year')
plt.ylabel('Per capita Income')
plt.show()
```

```python
#Visualize test dataset result
plt.scatter(x_test,y_test,color='blue')  #Actual values
plt.plot(x_test,lr.predict(x_test),color='red') #Predict value using linear regression
plt.title('To find Per Capita Income')
plt.xlabel('year')
plt.ylabel('Per capita Income')
plt.show()
```

```python
#Calculate residuals
from sklearn import metrics
#Mean Absolute Error
print('MAE',metrics.mean_absolute_error(y_test,y_pred))
#Mean square error
print("MSE",metrics.mean_squared_error(y_test,y_pred))
#Root Mean Squared error
print("RMSE",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
